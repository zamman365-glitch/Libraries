# Q1. Select Single Column

## Problem Statement
Create a DataFrame and display only the Name column.

## Sample Input
```python
Name, Age
```

## Sample Output
```python
Only Name column
```


````python
# Write your code here
import pandas as pd
data=pd.DataFrame({
    'Name':['aman','bilal','LL','Goti'],
    'Age':[20,21,7,21],
    'Marks':[90,100,79,50]
})
print(data['Name'])
````


# Q2. Select Multiple Columns

## Problem Statement

## Sample Input
```python
Name, Marks
```

## Sample Output
```python
Selected columns
```


````python
# Write your code here
print(data[['Name','Marks']])
````


# Q3. Filter Students


# Q4. Filter by City

## Problem Statement
Display students belonging to Bhopal.

## Sample Input
```python
City = Bhopal
```

## Sample Output
```python
Matching rows
```


````python
# Write your code here
df = pd.DataFrame({
    "Name": ["Aman", "Bilal", "Rahul", "Sara"],
    "City": ["Bhopal", "Indore", "Bhopal", "Delhi"],
    "Marks": [85, 90, 78, 92]
})


print(df[df["City"] == "Bhopal"])
````

# Seaborn – Class 2: Categorical Plots


## What Are Categorical Plots?

Categorical plots visualize data where at least one variable is **categorical**  
(e.g., day of week, gender, class).

Seaborn gives several easy options to:
- Compare **counts** or **averages** across categories
- Understand spread & distribution


## barplot() vs countplot()

```python
import seaborn as sns

# Load one of Seaborn's built-in practice datasets
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.barplot(x=tips['day'],y=tips['total_bill'])  # shows the average value of a numeric column for each category, with error bars.
# mostly use in ML for class imbalance
```

```python
sns.countplot(x=tips['day'])  # kitne frequently cheezen a rahe hai
```

## boxplot()

```python
sns.boxplot(x=tips['sex'],y=tips['total_bill'])  # one of the most imp plot
```

## stripplot() & swarmplot()

```python
sns.stripplot(x=tips['day'],y=tips['total_bill'],hue=tips['sex'])
```

```python
sns.swarmplot(x=tips['day'],y=tips['total_bill'],hue=tips['sex'])#
```

| Plot        | Best for |
|-------------|-----------|
| **barplot** | Compare average values across categories |
| **countplot** | Show how many items fall in each category |
| **boxplot** | See median, quartiles, and outliers |
| **violinplot** | Same as boxplot + shows distribution |
| **stripplot** | View every observation (good for small data) |
| **swarmplot** | Like stripplot but avoids overlap |


```python
#1) how many smokers vs no smoker group by gender
#2) average total bill accounding to smoker
#3) average tip by day , group by time
#4) spread of total bill across days
#5) tip distribution by smoker , group by gender
```

```python
#1) how many smokers vs no smoker group by gender
sns.countplot(x=tips['smoker'],hue=tips['sex'])
```

```python
#2) average total bill accounding to smoker
sns.barplot(x=tips['smoker'],y=tips["total_bill"])
```

```python
#3) average tip by day , group by time
sns.barplot(x=tips['day'],y=tips['tip'],hue=tips['time'])
```

```python
#4) spread of total bill across days
sns.boxplot(x=tips['day'],y=tips['total_bill'])
```

```python
#5) tip distribution by smoker , group by gender
sns.boxplot(x=tips['smoker'],y=tips['tip'],hue=tips['sex'])
```

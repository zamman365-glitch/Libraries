## 1 Inspecting Data
Quick ways to understand your dataset:
- `head(n)` ŌåÆ first *n* rows (default 5)
- `tail(n)` ŌåÆ last *n* rows
- `info()` ŌåÆ columns, non-null counts, and dtypes
- `shape` ŌåÆ (rows, columns)
- `dtypes` ŌåÆ data type of each column

```python
import pandas as pd

# Create a small DataFrame
a = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Age": [25, 30, 35, 40, 22],
    "Score": [88, 92, 79, 85, 95]
}

b = pd.DataFrame(a)


#3. .inof()
print(b.info())

#4. .
```

## 2 Selecting Columns & Rows
Ways to access parts of a DataFrame:
- Single column ŌåÆ `df['col']`
- Multiple columns ŌåÆ `df[['col1','col2']]`
- Label-based selection ŌåÆ `df.loc[row_label, col_label]`
- Integer-index selection ŌåÆ `df.iloc[row_idx, col_idx]`
Tip: `:` means "all rows/columns" in that dimension.

```python
print(b)

#1. Single column
print(b['Age'])

#2. Multiple columns
print(b[['Age','Score']])


#3. Select only row index 2 and column score
print(b.loc[2,'Score'])
print(b.loc[2,['Age','Score']]) #multiple column


#4. select using indexing we use .iloc()
print(b.iloc[0:3,0:2])




```

## 3 Index & Columns Overview
- `df.index` ŌåÆ row labels
- `df.columns` ŌåÆ column names
- `df.values` ŌåÆ underlying NumPy array (read-only; prefer using DataFrame ops)

```python
#1. .index()
print(b.index) # rows kaha se start ho rahe hai kahan stop ho rahe hai

#2..columns-> name of the columns
print(b.columns)

#3. .values -> dataframe convert into matrix
print(b.values)
```

## 4 Renaming Columns & Setting an Index
- Rename columns ŌåÆ `df.rename(columns={'old':'new'}, inplace=False)`
- Set a column as index ŌåÆ `df.set_index('column_name', inplace=False)`
- Reset index ŌåÆ `df.reset_index(inplace=False)`
 If you don't use `inplace=True`, remember to assign the result back to a variable.

```python
print(b)

#1. .rename(old_name:new_name)
#inplace-> Original dataframe mai changes apply kar deta hai.
b.rename(columns={'Score':'Marks'},inplace=True)
print(b)

#2. set index  important hota hai bht
c=b.set_index('Name')
print(c)

#3. .reset index
c.reset_index(inplace=True)
print(c)

```

## 5 Basic Attributes Recap
- `df.columns` ŌåÆ names of all columns
- `df.index` ŌåÆ row labels
- `df.values` ŌåÆ NumPy array (data only)
These are handy for quick checks and loops (though vectorized ops are preferred).

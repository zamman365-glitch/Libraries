```python
#Import matplotlib
import matplotlib.pyplot as plt
```

```python
#Line plot.
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
```

```python
plt.plot(x, y)

plt.xlabel('This is X-Label')
plt.ylabel('This is Y-Label')
plt.title('This is our first Line Plot')

plt.show()
```

```python
plt.figure(figsize=(6,4))

x  = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]


#line 1 plot
plt.plot(x, y1, label = "Line 1", color="pink")
plt.plot(x, y2, label = "Line 2", color="red")

plt.grid()
plt.legend()
```

```python
x  = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

plt.subplot(2,2,1)
plt.plot(x,y1,'r--')
plt.xlabel('This is X-Label')
plt.ylabel('This is Y-Label')
plt.title('This is our first section Plot')
plt.grid()

plt.subplot(2, 2, 4)
plt.plot(x,y2,'g')
plt.xlabel('This is X-Label')
plt.ylabel('This is Y-Label')
plt.title('This is our fourth section Plot')
plt.grid()


plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

fig = plt.figure(figsize=(6,4))

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #([left,bottom,width,height])
ax1.plot(x,y)
ax1.set_xlabel("X-axis")
ax1.set_ylabel("Y-axis")
plt.title("plot1")



ax2 = fig.add_axes([0.2,0.5,0.3,0.3])
ax2.plot(x,y)
plt.title("plot2")



plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure(figsize=(8, 4), dpi=1000)  #dpi=controls zoom in or zoom out for a graph
plt.title("Custom Figure Size & DPI")

plt.plot(x, y)
plt.grid(True)
```

```python
x  = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

plt.plot(x, y1, label = "Line 1", color="black",lw=2,marker='o')
plt.plot(x, y2, label = "Line 2", color="red",lw=1,ls='-.',marker='*')

plt.legend()


plt.show()
```

```python
#histplot -> Shows spread of only one column/varibles
import numpy as np
nums=np.random.randn(100)
plt.hist(nums,bins=20)
```

```python
#box plot -> shows outliers of a single column
plt.boxplot(nums)
#min
#25%  ->lower quatile
#50%  -> median
#75%  -> Upper quatile
#max

# Values between lower and upper quatile are genric or actual values

# Values outside of lower and upper quatile are considered to be an outliers

```

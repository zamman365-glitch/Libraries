# Q1. Fancy Indexing (1D Array)

## Problem Statement
Create the array:

```python
[10, 20, 30, 40, 50, 60]
```

Using fancy indexing, print:
- 10
- 30
- 60

## Sample Output
```python

```


# Q4. Conditional Selection

## Problem Statement
Create the array:

```python
[5, 12, 18, 25, 30, 7]
```

Print only values greater than 15.

## Sample Output
```python
[18 25 30]
```


````python
# Write your code here
arr=np.array([5, 12, 18, 25, 30, 7])
print(np.where(arr>15,arr,0))
````


# Q5. Conditional Selection

## Problem Statement
Create the array:

```python
[2, 4, 7, 10, 13, 16]
```

Print only even numbers.

## Sample Output
```python
[ 2  4 10 16]
```


````python
import numpy as np

arr = np.array([2, 4, 7, 10, 13, 16])

print(arr[arr % 2 == 0])
````


# Q9. Vertical Stacking

## Problem Statement
Create two arrays:

```python
[1,2,3]
[4,5,6]
```

Combine them using vertical stacking.

## Sample Output
```python
[[1 2 3]
 [4 5 6]]
```


````python
# Write your code here
a=np.array([1,2,3])
b=np.array([4,5,6])
np.vstack([a,b])
````


# Q10. Horizontal Stacking

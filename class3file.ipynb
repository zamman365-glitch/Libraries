##Numpy array operations

Lets first see how arrays behave like “vectors/matrices” when operations happen element-wise.

```python
import numpy as np

a=np.array([10,20,30,40])
b=np.array([1,2,3,4])

#1 Addition of the array
print(a+b)

#2 subtraction the array
print(a-b)

#3 Multiplication of the arrays
print(a*b)

```

These things are already different from List and if we talk about one more special power it is **Broadcasting**

```python
a=np.array([10,20,30,40])
print(a*5)
```

Even these things work with different shaped array the calculations will happen on every row. but dimensions must match

```python
A=np.array([[10,20,30],[40,50,60]])
B=np.array([10,20,30])
print(A+B)
```

#Aggregate Functions

These functions are used to calculate the statistics on arrays.

```python
# Basic Statistical operations perform
arr=np.array([4,78,21,34,100,56])

#1. sum
print(f"Sum is : {arr.sum()}")

#2. Mean
print(f"Mean is : {arr.mean()}")


#3. Min
print(f"Min is : {arr.min()}")

#4. Max
print(f"Max is : {arr.max()}")

#5. Standard deviation
print(f"Standard deviation : {arr.std()} ")

#6. Varaince
print(f"Varaince : {arr.var()}")

```

You can also do the axis based operations row-Wise and Colum-wise

```python
data=np.array([[10,20,30],[40,50,60],[70,80,90]])
print(f"Sum of all : { data.sum()}")
#axis=0 -> columns
#axis=1 -> rows
print(f'Sum of columns:{np.sum(data,axis=0)}')
print(f'Sum of rows:{np.sum(data,axis=1)}')
```
